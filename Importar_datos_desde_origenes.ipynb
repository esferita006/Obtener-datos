```python
import pandas as pd
import wget
from google.colab import files, drive
import gspread
from oauth2client.client import GoogleCredentials

# Función para generar un DataFrame desde un archivo CSV en línea
def create_dataframe_from_web():
    url = input("Ingrese la URL del archivo CSV en línea: ")
    filename = wget.download(url)
    df = pd.read_csv(filename)
    return df

# Función para generar un DataFrame desde un archivo CSV proporcionado por el usuario
def create_dataframe_from_upload():
    uploaded = files.upload()
    filename = list(uploaded.keys())[0]
    df = pd.read_csv(filename)
    return df

# Función para generar un DataFrame desde un archivo CSV en Google Drive
def create_dataframe_from_drive():
    drive.mount('/content/drive')
    file_path = input("Ingrese la ruta del archivo CSV en Google Drive: ")
    df = pd.read_csv(file_path)
    return df

# Función para generar un DataFrame desde Google Spreadsheets
def create_dataframe_from_google_spreadsheets():
    auth.authenticate_user()
    gc = gspread.authorize(GoogleCredentials.get_application_default())
    spreadsheet_url = input("Ingrese la URL de Google Spreadsheets: ")
    sheet_name = input("Ingrese el nombre de la hoja: ")
    worksheet = gc.open_by_url(spreadsheet_url).worksheet(sheet_name)
    data = worksheet.get_all_values()
    df = pd.DataFrame(data[1:], columns=data[0])
    return df

# Menú interactivo
while True:
    print("Menú Principal:")
    print("1. Generar DataFrame desde archivo CSV en línea")
    print("2. Generar DataFrame desde archivo CSV proporcionado por el usuario")
    print("3. Generar DataFrame desde archivo CSV en Google Drive")
    print("4. Generar DataFrame desde Google Spreadsheets")
    print("5. Salir")

    option = input("Seleccione una opción (1/2/3/4/5): ")

    if option == '1':
        df = create_dataframe_from_web()
        print("DataFrame generado con éxito.")
    elif option == '2':
        df = create_dataframe_from_upload()
        print("DataFrame generado con éxito.")
    elif option == '3':
        df = create_dataframe_from_drive()
        print("DataFrame generado con éxito.")
    elif option == '4':
        df = create_dataframe_from_google_spreadsheets()
        print("DataFrame generado con éxito.")
    elif option == '5':
        break
    else:
        print("Opción no válida. Por favor, seleccione una opción válida.")

print("Programa terminado.")
```

```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import requests
from io import StringIO

# Función para descargar y cargar un archivo CSV en un DataFrame
def load_csv_from_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            csv_data = StringIO(response.text)
            df = pd.read_csv(csv_data)
            return df
        else:
            return None
    except Exception as e:
        return None

# Función para manejar la selección del usuario
def handle_selection(change):
    selection = change.new
    if selection == "Generar DataFrame desde URL":
        url = input_url.value
        df = load_csv_from_url(url)
        if df is not None:
            print("DataFrame cargado exitosamente:")
            display(df)
        else:
            print("No se pudo cargar el DataFrame desde la URL proporcionada.")
    elif selection == "Salir":
        print("¡Adiós!")
    else:
        print("Selección no válida.")

# Crear el menú interactivo
menu_options = ["Generar DataFrame desde URL", "Salir"]
menu = widgets.Dropdown(
    options=menu_options,
    description="Seleccione una opción:",
    style={"description_width": "initial"},
)

# Crear un campo de entrada para la URL
input_url = widgets.Text(
    description="Ingrese la URL del archivo CSV:"
)

# Agregar un botón para ejecutar la acción
button = widgets.Button(
    description="Cargar DataFrame"
)

# Asociar la función handle_selection al cambio en la selección del menú
menu.observe(handle_selection, names='value')

# Definir la función para cargar el DataFrame al hacer clic en el botón
def load_dataframe(button):
    url = input_url.value
    df = load_csv_from_url(url)
    if df is not None:
        print("DataFrame cargado exitosamente:")
        display(df)
    else:
        print("No se pudo cargar el DataFrame desde la URL proporcionada.")

# Asociar la función load_dataframe al evento de clic en el botón
button.on_click(load_dataframe)

# Mostrar el menú interactivo y los widgets
display(menu)
display(input_url)
display(button)
```
